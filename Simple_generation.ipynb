```python
##two objcet same color same size
import json
with open("data/properties.json", 'r') as f:
    properties = json.load(f)
shapes=list(properties["shapes"].keys())
#shapes=["cone","pyramid"]
for shape1 in shapes:
    for shape2 in shapes:
        if shape1==shape2:
            continue
        color="blue"
        arg_shapes=shape1+" "+shape2
        arg_colors=color+" "+color
        arg_rotates="0,0,0 0,0,0"
        arg_scales="1 1"
        arg_locs="-2.5,2.5 2.5,-2.5"
        !../blender-4.1.1-linux-x64/blender --background --python render_images_custom.py -- --use_gpu 1  --objects {arg_shapes}\
             --cobjects {arg_colors} --robjects 0 0 --sobjects 1 1 --objs_config data/TwoObjectSameColor.json  --num_images 1\
             --margin 0.01 --width 1024 --height 1024 --output_image_dir ../output/Two-Object-Same-Color/
```
